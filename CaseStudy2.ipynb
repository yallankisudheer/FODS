```python
import pandas as pd
file_path = '/content/payroll_dataset.xlsx'
xls = pd.ExcelFile(file_path)

```

```python
# Merge the Employees and Departments tables and display the first 5 rows of the
# merged dataset.
employees_df = pd.read_excel(xls, sheet_name='Employees')
departments_df = pd.read_excel(xls, sheet_name='Departments')
merged_df = pd.merge(employees_df, departments_df, on='DepartmentID')
merged_df.head()
```

```python
cleaned_df = merged_df.drop_duplicates()
rows_before = merged_df.shape[0]
rows_after = cleaned_df.shape[0]

rows_before, rows_after
```

```python
# Explain and apply the interquartile range (IQR) method to detect and handle outliers in
# the Salary column.
Q1 = cleaned_df['Salary'].quantile(0.25)
Q3 = cleaned_df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_removed_df = cleaned_df[(cleaned_df['Salary'] >= lower_bound) & (cleaned_df['Salary'] <= upper_bound)]
rows_before_outlier_removal = cleaned_df.shape[0]
rows_after_outlier_removal = outliers_removed_df.shape[0]

lower_bound, upper_bound, rows_before_outlier_removal, rows_after_outlier_removal
```

```python
# Apply one-hot encoding to the Location column and show the updated dataset.
one_hot_encoded_df = pd.get_dummies(cleaned_df, columns=['Location'])
one_hot_encoded_df
```

```python
# Perform Min-Max normalization on the Salary column and display the transformed
# dataset.
min_salary = cleaned_df['Salary'].min()
max_salary = cleaned_df['Salary'].max()
cleaned_df['Salary_Normalized'] = (cleaned_df['Salary'] - min_salary) / (max_salary - min_salary)
cleaned_df.head()
```
