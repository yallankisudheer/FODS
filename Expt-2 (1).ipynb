```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("D:\\Data Science\\Datasets\\housing_dataset.csv")
df
```

```python
null_col=df.columns[df.isnull().any()]
null_col
```

```python
med_price=df['Price'].median()
df['Price'].fillna(med_price)
```

```python
mean_bath=df['Bathrooms'].mean()
df['Bathrooms'].fillna(mean_bath)
```

```python
df.dropna(inplace=True)
```

```python
Q1=df['Price'].quantile(0.25)
Q2=df['Price'].quantile(0.75)
```

```python
IQR=Q2-Q1
lb=Q1 - 1.5 * IQR
ub=Q2 + 1.5 *IQR
```

```python
Outliers=df[(df['Price']>=lb) | (df['Price']<=ub)]
Outliers
```

```python
scaler=MinMaxScaler()
df['Sqft_Living']=scaler.fit_transform(df[['Sqft_Living']])
df
```

```python
df['Year_Built']=df['Year_Built'].astype(int)
```

```python
df['Waterfront']=df['Waterfront'].apply(lambda x:1 if x=='Yes' else 0)
```

```python
df.to_csv('Cleaned_dataset.csv',index=False)
```
